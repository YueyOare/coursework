Заполнение пропусков

```python
import pandas as pd
import numpy as np
from fancyimpute import IterativeImputer
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor

def impute_by_mean_variate(col):
    mean = col.mean()
    std = col.std()
    random_values = np.random.uniform(mean - std, mean + std, size=len(col))
    return col.combine_first(pd.Series(random_values, index=col.index))


def impute_by_value(data_, method='constant', value=0):
    if isinstance(data_, pd.Series):
        # Если входной объект - Series
        imputer = SimpleImputer(missing_values=np.nan, strategy=method, fill_value=value)
        imputed_data = imputer.fit_transform(data_.values.reshape(-1, 1))
        return pd.Series(imputed_data.flatten(), index=data_.index)
    elif isinstance(data_, pd.DataFrame):
        # Если входной объект - DataFrame
        imputer = SimpleImputer(missing_values=np.nan, strategy=method, fill_value=value)
        imputed_data = imputer.fit_transform(data_)
        return pd.DataFrame(imputed_data, columns=data_.columns)
    else:
        raise ValueError("Unsupported input type. Supported types are pd.Series and pd.DataFrame.")


def impute_by_norm_dist(col, norms=None):
    # col.astype('Float64')
    # print(col.dtype)
    size = len(col)

    if norms is not None:
        # Учитываем референсные значения (нормы)
        mean = (norms[1] + norms[0]) / 2
        std = (norms[1] - norms[0]) / 6
    else:
        mean = col.mean()
        std = col.std()

    random_values = np.random.normal(loc=mean, scale=std, size=size)

    return col.combine_first(pd.Series(random_values, index=col.index))


def impute_by_imputer(data_):
    imputer = IterativeImputer(random_state=0, estimator=RandomForestRegressor(
        # We tuned the hyperparameters of the RandomForestRegressor to get a good
        # enough predictive performance for a restricted execution time.
        n_estimators=4,
        max_depth=10,
        bootstrap=True,
        max_samples=0.5,
        n_jobs=-1,
        random_state=0,
    ), max_iter=25, verbose=1)
    imputed_data = imputer.fit_transform(data_)
    return data_.combine_first(pd.DataFrame(imputed_data, columns=data_.columns))
```

```python
male_data = data_imputed[data_imputed['Пол'] == 1]
female_data = data_imputed[data_imputed['Пол'] == 2]
```

```python
common_normal_data = data_imputed[list(normal_columns_df['Нормальные в общем'].dropna())]

male_normal_data = data_imputed[list(normal_columns_df['Нормальные (м)'].dropna())]

female_normal_data = data_imputed[list(normal_columns_df['Нормальные (ж)'].dropna())]
```

Заполнение числовых переменных

```python
for column in data_imputed.columns:
    if column in norms_df.iloc[:, 0].values:
        # Извлекаем соответствующие нормы
        norms = norms_df[norms_df.iloc[:, 0] == column]
        if column in female_normal_data.columns:
            female_data[column] = impute_by_norm_dist(female_normal_data[column], norms[['Минимум (ж)', 'Максимум (ж)']].values[0])

        if column in male_normal_data.columns:
            male_data[column] = impute_by_norm_dist(male_normal_data[column], norms[['Минимум (м)', 'Максимум (м)']].values[0])

        if column in common_normal_data.columns:
            data_imputed[column] = impute_by_norm_dist(common_normal_data[column], norms[['Минимум (ж)', 'Максимум (ж)']].values[0])
    else:
        if column in female_normal_data.columns:
            female_data[column] = impute_by_norm_dist(female_normal_data[column])

        if column in male_normal_data.columns:
            male_data[column] = impute_by_norm_dist(male_normal_data[column])

        if column in common_normal_data.columns:
            data_imputed[column] = impute_by_norm_dist(common_normal_data[column])
```

```python
data_imputed[data_imputed['Пол'] == 1] = male_data
data_imputed[data_imputed['Пол'] == 2] = female_data
```
